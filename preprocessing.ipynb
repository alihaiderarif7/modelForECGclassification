```python
from PIL import Image
import os


# Input and output folder paths
input_folder = '/content/gdrive/MyDrive/ecg_images/normal'
output_folder = '/content/gdrive/MyDrive/croped_images/normal'

# Create the output folder if it doesn't exist
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# List all files in the input folder
input_files = os.listdir(input_folder)

# Loop through each file in the input folder
for file_name in input_files:
    # Check if the file is an image (you can add more image file extensions if needed)
    if file_name.lower().endswith(('.png', '.jpg', '.jpeg', '.gif', '.bmp')):
        # Construct the full path of the input image
        input_image_path = os.path.join(input_folder, file_name)

        # Open the image file
        img = Image.open(input_image_path)

        # Specify the coordinates for cropping (top-left, bottom-right)
        x1, y1 = 83, 300  # Top-left corner
        x2, y2 = 2151, 1500  # Bottom-right corner

        # Crop the image
        cropped_img = img.crop((x1, y1, x2, y2))

        # Construct the full path for the output cropped image
        output_image_path = os.path.join(output_folder, f'cropped_{file_name}')

        # Save the cropped image to the output folder
        cropped_img.save(output_image_path)

# End of loop
```

```python
import os
import cv2
import numpy as np

# Input and output folder paths
input_folder = '/content/gdrive/MyDrive/croped_images/normal'
output_folder = '/content/gdrive/MyDrive/gridRemoval/normal'

# Create the output folder if it doesn't exist
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# List all files in the input folder
input_files = os.listdir(input_folder)

# Loop through each file in the input folder
for file_name in input_files:
    # Check if the file is an image (you can add more image file extensions if needed)
    if file_name.lower().endswith(('.png', '.jpg', '.jpeg', '.gif', '.bmp')):
        # Construct the full path of the input image
        input_image_path = os.path.join(input_folder, file_name)

        # Read the image using OpenCV
        ECG_image = cv2.imread(input_image_path)

        # Convert the image to grayscale
        img_gray = cv2.cvtColor(ECG_image, cv2.COLOR_BGR2GRAY)

        # Apply Canny edge detection
        img_edges = cv2.Canny(img_gray, 800, 800, apertureSize=3)

        # Construct the full path for the output edge-detected image
        output_image_path = os.path.join(output_folder, f'remove_grid_{file_name}')

        # Save the edge-detected image to the output folder
        cv2.imwrite(output_image_path, img_edges)

# End of loop
```
